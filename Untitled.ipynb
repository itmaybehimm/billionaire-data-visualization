# Billionaires Statistics Dataset

The "Billionaires Statistics Dataset" is a comprehensive collection of data providing valuable insights into the world of billionaires. This dataset includes a wide range of information about individuals who have amassed significant wealth, offering a glimpse into their wealth distribution, the industries they are involved in, their personal demographics, and the factors contributing to their financial success.

## Key Features

- **rank:** The ranking of the billionaire in terms of wealth.
- **finalWorth:** The final net worth of the billionaire in U.S. dollars.
- **category:** The category or industry in which the billionaire's business operates.
- **personName:** The full name of the billionaire.
- **age:** The age of the billionaire.
- **country:** The country in which the billionaire resides.
- **city:** The city in which the billionaire resides.
- **source:** The source of the billionaire's wealth.
- **industries:** The industries associated with the billionaire's business interests.
- **countryOfCitizenship:** The country of citizenship of the billionaire.
- **organization:** The name of the organization or company associated with the billionaire.
- **selfMade:** Indicates whether the billionaire is self-made (True/False).
- **status:** "D" represents self-made billionaires (Founders/Entrepreneurs) and "U" indicates inherited or unearned wealth.
- **gender:** The gender of the billionaire.
- **birthDate:** The birthdate of the billionaire.
- **lastName:** The last name of the billionaire.
- **firstName:** The first name of the billionaire.
- **title:** The title or honorific of the billionaire.
- **date:** The date of data collection.
- **state:** The state in which the billionaire resides.
- **residenceStateRegion:** The region or state of residence of the billionaire.
- **birthYear:** The birth year of the billionaire.
- **birthMonth:** The birth month of the billionaire.
- **birthDay:** The birth day of the billionaire.
- **cpi_country:** Consumer Price Index (CPI) for the billionaire's country.
- **cpi_change_country:** CPI change for the billionaire's country.
- **gdp_country:** Gross Domestic Product (GDP) for the billionaire's country.
- **gross_tertiary_education_enrollment:** Enrollment in tertiary education in the billionaire's country.
- **gross_primary_education_enrollment_country:** Enrollment in primary education in the billionaire's country.
- **life_expectancy_country:** Life expectancy in the billionaire's country.
- **tax_revenue_country_country:** Tax revenue in the billionaire's country.
- **total_tax_rate_country:** Total tax rate in the billionaire's country.
- **population_country:** Population of the billionaire's country.
- **latitude_country:** Latitude coordinate of the billionaire's country.
- **longitude_country:** Longitude coordinate of the billionaire's country.

You can access this dataset for further analysis and exploration on Kaggle using the provided [link](https://www.kaggle.com/datasets/nelgiriyewithana/billionaires-statistics-dataset).


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data= pd.read_csv("Billionaires-Statistics-Dataset.csv")
data.head()
```

```python
#for our purposes, we won't be needing some columns we drop them below
#city can be neglected as billanires are cpuntry level rich
#there is no clear meaning provided to different values of status
del_cols=['birthDate','lastName','firstName','date','birthYear','birthMonth','city','birthDay','latitude_country','longitude_country','organization','status','state','residenceStateRegion']
data.drop(del_cols,axis=1,inplace=True)
data.info()
```

```python
data.head()
```

```python
data.isnull().sum()
```

## Observation

- for null values of country, we can assume the country of citzenship to be the country of birth

- The "country" column has fewer missing values (38) compared to several related columns, including "cpi_country," "gdp_country," "tax_revenue_country_country," "life_expectancy_country," "total_tax_rate_country," and "population_country."

These columns related to economic and demographic factors in the billionaire's country have a higher number of missing values (ranging from 164 to 184) when compared to the "country" column. It's worth noting that the "country" column provides information about the billionaire's country of residence, while the other columns offer contextual data about the economic and demographic conditions in those countries.

we can assume that for same countries, these values must be same

```python
data['country'].fillna(data['countryOfCitizenship'],inplace=True)
#white sapces at end remove
data['country'] = data['country'].str.strip()
```

```python
country_gdp_group=data.groupby(['countryOfCitizenship','gdp_country'])

for name, entries in country_gdp_group:
    print(f"{name[0]}:{name[1]}")
```

# Observation
we can see over here that the gdp_data are not consistent so let's use another csv file to populate it

```python
#so we drop and remake gdp column
data.drop('gdp_country',axis=1,inplace=True)
```

```python
gdp_data = pd.read_csv('gdp-data.csv')
gdp_data.head()
```

```python
#we need only the latest gdp i.e 2022, so we drop all columns except Country Name and 2022
remove_cols=gdp_data.columns.to_list()
remove_cols.remove('Country Name')
remove_cols.remove('2022')
gdp_data.drop(remove_cols,axis=1,inplace=True)
gdp_data.info()
```

```python
#we add gdp_country from new file
#it seems some countries have different names so we adjust
replace_dict={"Hong Kong SAR, China":"Hong Kong","Russian Federation":"Russia","Czechia":"Czech Republic","Korea, Rep.":"South Korea","Egypt, Arab Rep.":"Egypt","Slovak Republic":"Slovakia","Eswatini":"Eswatini (Swaziland)"}
gdp_data.replace(replace_dict,inplace=True)
country_to_gdp = gdp_data.set_index('Country Name')['2022'].to_dict()
```

```python
data['gdp_country'] = data['country'].map(country_to_gdp)
#remove rows whose gdp not present in above table
data.dropna(subset = ['gdp_country'], inplace=True)
```

```python
tax_data = pd.read_csv('tax-data.csv')
remove_cols=tax_data.columns.to_list()
remove_cols.remove('country')
remove_cols.remove('rate')
tax_data.drop(remove_cols,axis=1,inplace=True)
tax_data.head()
```

```python
replace_dict={"China, Hong Kong Special Administrative Region":"Hong Kong","Russian Federation":"Russia","Czechia":"Czech Republic","Republic of Korea":"South Korea","Egypt, Arab Rep.":"Egypt","Slovak Republic":"Slovakia","Eswatini":"Eswatini (Swaziland)","United States of America":"United States","United Kingdom of Great Britain and Northern Ireland":"United Kingdom"}
tax_data.replace(replace_dict,inplace=True)
country_to_tax = tax_data.set_index('country')['rate'].to_dict()
```

```python
data['tax_country'] = data['country'].map(country_to_tax)
data.dropna(subset = ['tax_country'], inplace=True)
```

# We can do the same things to other colums but we will just drop the columns for this report

```python
rem_col=['cpi_country','cpi_change_country','gross_tertiary_education_enrollment','gross_primary_education_enrollment_country','life_expectancy_country','tax_revenue_country_country','total_tax_rate_country','population_country']
data.drop(rem_col,axis=1,inplace=True)
```

```python
#self made can be represented by 1 or 0
data['selfMade'].replace({False: 0, True: 1}, inplace=True)
```

```python
data['gender'].value_counts()
#hence we can represent M by 0 and F by 1
```
